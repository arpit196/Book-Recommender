```python
col_names=['author','books']
df2=pd.read_csv('C:/Users/Ishu/Desktop/books1.txt',delim_whitespace=False,delimiter=',',names=col_names)
```

```python
from scipy.stats.stats import pearsonr  
cor=np.zeros((54,54))
for i in range(0,53) :
    for j in range(0,53) :
        if not i == j :
            list1=[]
            list2=[]
            for k in range(0,85) :
                if (not df1.iloc[i,k]==0) and (not df1.iloc[j,k]==0) :
                    list1.append(df1.iloc[i,k])
                    list2.append(df1.iloc[j,k])
            arr1=np.asarray(list1)
            arr2=np.asarray(list2)
            cor[i][j]=pearsonr(arr1,arr2)[0];
            if np.isnan(cor[i][j]) :
                cor[i][j]=0
```
